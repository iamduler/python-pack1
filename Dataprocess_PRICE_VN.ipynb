```python
# Kết nối Google Drive với Colab
from google.colab import drive
drive.mount('/content/drive')

# Nhập thư viện pandas để xử lý dữ liệu
import pandas as pd

# Đường dẫn tệp
file_path = '/content/drive/MyDrive/Cleaned data/Cleaned_Vietnam_Price.xlsx'

# Đọc danh sách các sheet trong tệp
excel_file = pd.ExcelFile(file_path)
sheets = excel_file.sheet_names  # Danh sách tên sheet
print(f"Danh sách các sheets: {sheets}")

# Đọc dữ liệu từ từng sheet
sheet1 = pd.read_excel(file_path, sheet_name='Sheet1')
sheet2 = pd.read_excel(file_path, sheet_name='Sheet2')

# Hiển thị vài dòng đầu của từng sheet

print("\nSheet: Sheet1")
print(sheet1.head())

print("\nSheet: Sheet2")
print(sheet2.head())
```

```python
sheet1
```

```python
# Lọc các dòng có lỗi (#ERROR hoặc $$ER)
error_rows = sheet2[sheet2.apply(lambda x: x.astype(str).str.contains(r'#ERROR|\$\$ER', regex=True, na=False)).any(axis=1)]

# Hiển thị số dòng lỗi
print(f"Số dòng bị lỗi: {len(error_rows)}")

# Hiển thị các dòng lỗi
display(error_rows)

```

```python
sheet2 = sheet2[~sheet2.apply(lambda x: x.astype(str).str.contains('#ERROR|\$\$ER', regex=True, na=False)).any(axis=1)]
sheet2
```

```python
# Lọc các dòng có chứa từ "DEAD" trong cột 'Name'
dead_rows = sheet2[sheet2['Name'].str.contains('DEAD', na=False, case=False)]

# Hiển thị số dòng và các dòng bị lỗi
print(f"Số dòng chứa 'DEAD': {len(dead_rows)}")
display(dead_rows)
```

```python
# Xóa phần từ "DEAD" trở đi trong cột 'Name'
sheet2['Name'] = sheet2['Name'].str.split('DEAD').str[0]
sheet2
```

```python
# Lấy tên của hai cột cuối cùng
last_two_columns = sheet2.columns[-2:]  # Lấy tên 2 cột cuối
col1, col2 = last_two_columns[0], last_two_columns[1]

# Chuyển dữ liệu trong 2 cột cuối thành số thực
sheet2[col1] = pd.to_numeric(sheet2[col1], errors='coerce')
sheet2[col2] = pd.to_numeric(sheet2[col2], errors='coerce')

# Tính % thay đổi: (giá trị cột cuối - giá trị cột áp cuối) / giá trị cột áp cuối * 100
sheet2['% Change'] = ((sheet2[col2] - sheet2[col1]) / sheet2[col1]) * 100

# kiểm tra
sheet2[[col1, col2, '% Change']]
```

```python
sheet2['Code'] = sheet2['Code'].str.replace(r'\(P\)', '', regex=True)
sheet2
```

```python
# Danh sách cột cần lấy từ sheet1 qua sheet2
columns_to_merge =  ['Name', 'RIC', 'Start Date', 'Exchange', 'Sector', 'Activity']
sheet2 = pd.merge(sheet2, sheet1[columns_to_merge], on='Name', how='left', suffixes=('_sheet2', '_sheet1'))

sheet2
```

```python
# Lấy danh sách cột hiện tại
cols = sheet2.columns.tolist()

# Xác định các cột cần sắp xếp lại
columns_to_move = ['RIC', 'Start Date', 'Exchange', 'Sector', 'Activity']

# Đảm bảo các cột cần di chuyển có trong DataFrame (tránh lỗi nếu thiếu cột nào)
columns_to_move = [col for col in columns_to_move if col in cols]

# Loại bỏ các cột này khỏi danh sách ban đầu để tránh trùng lặp
remaining_cols = [col for col in cols if col not in ['Name', 'Code'] + columns_to_move]

# Xác định thứ tự cột mới
new_order = ['Name', 'Code'] + columns_to_move + remaining_cols

# Sắp xếp lại DataFrame theo thứ tự mới
sheet2 = sheet2[new_order]

# Hiển thị DataFrame sau khi sắp xếp
display(sheet2)
```

```python
import pandas as pd

# Đường dẫn để lưu file
output_file_path = "/content/drive/MyDrive/Cleaned data/Processed_Vietnam_Price.xlsx"

# Xuất sheet2 ra file Excel
sheet2.to_excel(output_file_path, sheet_name="Processed_Sheet2", index=False)

print(f"Dữ liệu đã được lưu vào: {output_file_path}")
```
